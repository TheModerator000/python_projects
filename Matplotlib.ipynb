```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

```python
x
```

```python
plt.plot(x,y,color='purple')
```

```python
#how to assign labels and a title to a plot
plt.plot(x,y,color='purple')
plt.xlabel('X label goes here')
plt.ylabel('Y label goes here')
plt.title('Title goes here')
```

```python
#(number of rows, number of columns, plot number)
plt.subplot(1,2,1)
plt.plot(x,y)
#assigning the second plot
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
#create plots using object oreinted method
fig = plt.figure()

#add set of axes to the figure
axes = fig.add_axes([0.8,1,0.8,0.8]) #[spcaing left, bottom, width, height] (values must be between 0 - 1)

#plot on axes
axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title goes here')
```

```python
#creating a figure with multiple axes
fig2 = plt.figure()

axes1 = fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
```

```python
#method that creates axes for you
fig, axes = plt.subplots()
axes.plot(x,y,'r')
```

```python
fig , axes = plt.subplots(nrows=1,ncols=2)

#to display your trend use the followig
axes[0].plot(x,y)
```

```python
fig , axes = plt.subplots(nrows=2,ncols=3)

#to plot on multidemsional figures
axes[0,0].plot(y,x,'r')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

#to create multiple figures with a for loop do this

for ax in axes:
    ax.plot(x,y,'g')
```

```python
#figure size and DPI

fig3 = plt.figure(figsize=(8,4),dpi=100)
```

```python
fig3,axes = plt.subplots(figsize=(12,3))

axes.plot(x,y)
```

```python
#saving plots to an image file
fig3.savefig('isds3200_graph.png', dpi=100)
```

```python
#legends to a plot
fig4 = plt.figure()
ax1 = fig4.add_axes([0,0,1,1])
ax1.plot(x,x**2, label = 'x**2')
ax1.plot(y,y**2, label = 'y**2')
ax1.legend(loc=2)
#1 upper right corner
#2 upper left corner
#3 lower left corner
#4 lower right corner
```

```python
#colors
fig4,ax = plt.subplots()
ax.plot(x,x**2,'g--')
ax.plot(x,x**3, 'r.-')
ax.plot(x,x**4,color='#461d7c')
```

```python
#modifying color, linewidth, and markers
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(x,x+1,color= '#461d7c', linewidth=2.00)
ax.plot(x,x+2,color='#461d7c', linewidth=.25)
ax.plot(x,x+3,color='green', linestyle='--', marker='+',markersize=10)
```

```python
#how to modify multiple aspects of a figure
fig,ax = plt.subplots(1,3,figsize=(12,6))

ax[0].plot(x,x**2,x,x**3)

ax[1].plot(x,x**2,x,x**3)
ax[1].set_ylim([0,60])
ax[1].set_xlim([2,5])

ax[2].plot(x,x**2,x,x**3)
ax[2].axis('tight')
```

```python
#how to create a scatter plot
plt.scatter(x,y)
```

```python
#how to create a histogram from random data
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
data1 = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data1, vert=True)
```

```python
#how to create grids and modify them
fig,axes = plt.subplots(1,2,figsize=(10,3))

axes[0].plot(x,x**2,x,x**3)
axes[0].grid(True)


axes[1].plot(x,x**2,x,x**3)
axes[1].grid(color='b', alpha=0.5,linestyle='dashed',linewidth=0.5)
```

```python
#how to create axes that span multiple columns and rows ((number of rows and cols), (position),col/rowspan)
fig6 = plt.figure()
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2)
ax3 = plt.subplot2grid((3,3),(1,2),rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,0))
ax5 = plt.subplot2grid((3,3),(2,1))

fig6.tight_layout()
```
