```python
import pandas as pd
import sqlite3 as sq3
```

```python
con = sq3.connect('chihook.db')
cur = con.cursor()
cur.execute('CREATE TABLE departments(Campus, Department, Unit, Year integer, Month integer, Day integer)')
con.commit
con.close()
```

```python
df = pd.read_csv('parsedData.csv')
df.head()
```

```python
df['Year'] = df['Created Year']
df['Month'] = df['Created Month']
df['Day'] = df['Created Day']

df.head()
```

```python
#dropping columns that are duplicates
df.drop('Created Year', axis=1,inplace = True)
df.drop('Created Month', axis=1,inplace = True)
df.drop('Created Day', axis=1,inplace = True)

df.head()
```

```python
con = sq3.connect('chihook.db')
df.to_sql('department', con)
con.commit()
con.close()
```
