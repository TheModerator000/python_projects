#### 18. Create a new column called "NewGenre". The content of this column should be populated from the data in column "Genre" by splitting the values in the column by space and extracting the last value in the list. (ex. country rap => rap) (10 pts)

```python
spotify['NewGenre'] = spotify['Genre'].apply(lambda title: title.split(' ')[-1])
spotify.head(3)
```
