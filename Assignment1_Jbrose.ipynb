### Create a function that counts the number of times a word occurs in another string

```python
def countWord(word, string2):
    list2 = string2.split()
    wordCount = 0 
    
    for w in list2:
        if w == word:
            wordCount += 1
    print(wordCount)
```

```python
countWord('wasted','Almost nothing was more annoying than having our wasted time wasted on something not worth wasting it on.')
```
