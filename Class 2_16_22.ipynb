```python
import numpy as np
missing_data = pd.DataFrame({'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]})
```

```python
missing_data
```

```python
missing_data.dropna()
```

```python
missing_data.dropna(axis=1)
```

```python
#if python finds that there are two instances of missing data it will drop that entity *by row*
missing_data.dropna(thresh=2)
```

```python
missing_data.fillna(value='dummy')
```

```python
#instead of getting rid of rows with null values, fill it with something like the mean of a column
missing_data['A'].fillna(value=missing_data['A'].mean())
```
