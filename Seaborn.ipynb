```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distplot

sns.displot(tips['total_bill'],kde=True)
```

```python
sns.displot(tips['total_bill'])
```

```python
#jointplot

sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') #kind: scatter,reg,resid,kde,hex
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
#pairplot
sns.pairplot(tips)
```

```python
#to seperate some of the data by a specific col
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
#catergory plots

#barplot and countplot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#boxplots and violnplots
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#stripplot and swarmplot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#factorplot

sns.factorplot(x='sex',y='total_bill',data=tips, kind='bar')
```

```python
#Matrix Plots
flights = sns.load_dataset('flights')
flights.head()
```

```python
#heatmaps (lighter the box the more the data is correlated)
sns.heatmap(tips.corr())
```

```python
#creating a pivot table with flights data
flights_pvt = flights.pivot_table(values='passengers', index='month', columns='year')
flights_pvt
```

```python
sns.heatmap(flights_pvt)
```

```python
#changing color of data points
sns.heatmap(flights_pvt,cmap='magma')
```

```python
sns.clustermap(flights_pvt)
```

```python
#grids and pairgrids

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
#styles and configs
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
#remove spine
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```
