```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df_left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df_right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
df_left_keys = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
df_right_keys = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df2
```

```python
df3
```

```python
#concatenating all of the records together just tacks em on at the end of the previous df
pd.concat([df1,df2,df3])
```

```python
df_concat = pd.concat([df1,df2,df3],axis=1)
```

```python
df_concat['A']
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_left,df_right,how='inner',on='key')
```

```python
df_left_keys
```

```python
df_right_keys
```

```python
pd.merge(df_left_keys,df_right_keys,how='inner',on=['key1','key2'])
```

```python
pd.merge(df_left_keys,df_right_keys,how='outer',on=['key1','key2'])
```

```python
pd.merge(df_left_keys,df_right_keys,how='left',on=['key1','key2'])
```

```python
pd.merge(df_left_keys,df_right_keys,how='right',on=['key1','key2'])
```

```python
left_j = pd.DataFrame({'A':['A0','A1','A2','A3'], 'B':['B0','B1','B2','B3']}, index=['k0','k1','k2','k3'])


right_j = pd.DataFrame({'C':['C0','C1','C2','C3'], 'D':['D0','D1','D2','D3']}, index=['k0','k1','k2','k3'])
```

```python
left_j
```

```python
right_j
```

```python
#join auto joins based on the indexes of a df
left_j.join(right_j)
```

```python
left_j.join(right_j,how='outer')
```
