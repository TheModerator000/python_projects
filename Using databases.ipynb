```python
con = sqlite3.connect('chinook.db')
```

```python
total_invoice_by_cust = pd.read_sql_query('SELECT CustomerId, SUM(Total) as Total FROM Invoices GROUP BY CustomerId',con)
total_invoice_by_cust.head(10)
```

```python
df_cust_empl = pd.read_sql_query('SELECT employees.EmployeeId, employees.LastName, employees.FirstName, customers.CustomerId FROM Employees INNER JOIN Customers on (supportrepid = employeeid)',con)

df_cust_empl.head(10)
```

```python
#joining tables in python
df_result = df_cust_empl.merge(total_invoice_by_cust,on='CustomerId', how='left')
```

```python
df_result.head(10)
```

```python
df_result.groupby(['LastName', 'FirstName'])['Total'].agg('sum')
```
